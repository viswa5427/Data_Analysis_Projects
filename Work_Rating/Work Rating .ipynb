```python
# import libraries
import numpy as np
import pandas as pd
```

```python
#Importing Data
df = pd.read_csv('wrk rating in quaranitne.csv' )
df.head()
```

```python
#size of Data
df.shape
```

```python
#converting WorK_Rate into float
df[["Work_Rate"]]=df[['Work_Rate']].astype("float")
```

```python
# checking the count of Nan values in each column in our Dataset
df.isnull().sum()
```

```python
# Replace Nan in Work_Rate with mean Value of Work_Rate
df["Work_Rate"].fillna(df["Work_Rate"].mean(),inplace=True)
```

```python
# cross checking of any Nan values in dataset is present or not!
df.isnull().sum()
```

```python
# we are using the inline backend
import matplotlib.pyplot as plt
```

```python
work=df[["Work_Rate"]]
work.head()
```

```python
Days=df[["Day"]]
Days.head()
```

```python
 #plot Days on X-axis,work_rate on Y-axis
plt.plot(Days,work)
plt.title('Work_Rating')
plt.ylabel('Work_Rate')
plt.xlabel('Day')
plt.ylim((0,5))#as work_rate range  is from 0-5 
plt.figure(figsize=(22, 30))
plt.show() 
```

```python
#Number of Days
Total_no_of_Days=Days.count()
print("Total_no_of_Days:",Total_no_of_Days[0]);
```

<p>Here in "Total_no_of_Days" we get all the details including column name , data type and the mean value
so inorder to access the mean frm that data we use Total_no_of_Days[0]</p>

```python
#It's the mean Work rate of 50 Days
mean=work.mean()
print(mean[0])
```
